```python
## Haxard rate function


# Defining the sequence and function
tau = 150
mdd = 300
z = 2
d = np.arange(0, 301, 10)

hazard = 1 - np.exp(-(d / tau) ** -z)

halfNormal = np.exp(-(d / tau) ** 2)

y = hazard

# Finding the point where the curve is closest to P(detection) = 0.5
closest_index = np.argmin(np.abs(y - 0.5))
edr_closest = d[closest_index]

# Finding the point where the curve flattens near zero
threshold = 0.8
mdd_index = np.where(y < threshold)[0][0]
mdd_value = d[mdd_index]

# Plotting the graph with the star
plt.plot(d, y, color='dodgerblue')
plt.scatter([edr_closest], [0.5], color='darkorange', marker='*', s=100)
plt.scatter([mdd_value], [y[mdd_index]], color='green', marker='*', s=100)

plt.ylim(0, 1)
plt.xlabel('Distance (m)')
plt.ylabel('P(detection)')
plt.title('Modified Hazard rate')
plt.annotate(f'EDR = {edr_closest:.0f} m', xy=(edr_closest, 0.5), xytext=(edr_closest-50, 0.6),
             arrowprops=dict(arrowstyle='->'))
plt.annotate(f'MDD = {mdd_value:.0f} m', xy=(mdd_value, y[mdd_index]), xytext=(mdd_value-50, y[mdd_index]+0.1),
             arrowprops=dict(arrowstyle='->'))
plt.show()
```

```python
# Defining the sequence and function
mdd = 300
z = 4
d = np.arange(0, 301, 10)
threshold = 0.02

# Defining function to plot curve and annotations
def plot_curve(tau, color):
    halfNormal = np.exp(-(d / tau) ** 2)
    y = halfNormal
    # Finding EDR and MDD
    edr_closest = d[np.argmin(np.abs(y - 0.5))]
    mdd_value = d[np.where(y < threshold)[0][0]]
    plt.plot(d, y, color=color)
    plt.scatter([edr_closest], [0.5], color=color, marker='*', s=100)
    plt.scatter([mdd_value], [y[np.where(y < threshold)[0][0]]], color=color, marker='*', s=100)
    plt.annotate(f'EDR = {edr_closest:.0f} m', xy=(edr_closest, 0.5), xytext=(edr_closest+10, 0.65),
                 arrowprops=dict(arrowstyle='->'))
    plt.annotate(f'MDD = {mdd_value:.0f} m', xy=(mdd_value, y[np.where(y < threshold)[0][0]]), xytext=(mdd_value-50, y[np.where(y < threshold)[0][0]]+0.1),
                 arrowprops=dict(arrowstyle='->'))

# Plotting both curves
plot_curve(tau=100, color='dodgerblue')
plot_curve(tau=150, color='darkorange')

plt.ylim(0, 1)
plt.xlabel('Distance (m)')
plt.ylabel('P(detection)')
plt.title('Half Normal')
plt.legend(['Tau = 100', 'Tau = 150'])
plt.show()
```

```python
# Absorption coefficients for lower and higher absorption
alpha_lower = 0.1  # dB/m
alpha_higher = 0.25  # dB/m
initial_dB = 100
d0 = 1  # Reference distance in meters

d = np.arange(0, 250, 10)

# Calculating sound level according to the inverse square law with absorption
L_d_lower_absorption = initial_dB - 10 * np.log10((d / d0) ** 2) - alpha_lower * d
edr_distance_lower = L_d_lower_absorption[np.where(d == 80)[0][0]]
print(edr_distance_lower)

L_d_higher_absorption = initial_dB - 10 * np.log10((d / d0) ** 2) - alpha_higher * d
edr_distance_higher = L_d_higher_absorption[np.where(d == 80)[0][0]]
print(edr_distance_higher)

# Plotting the graph for both lower and higher absorption
plt.figure(figsize=[10,6])

## lower absorption rate, lower attenuation, louder for longer
plt.plot(d, L_d_lower_absorption, color='darkorange',
         label='Lower Attenuation')

# label='Lower Absorption ( = 0.1 dB/m)')
plt.scatter([80], [edr_distance_lower], color='darkorange', marker='*', s=400)

## EDR value LOWER
plt.annotate(f'Minimum EDR = 80 m',
             xy=(80, edr_distance_lower),
             xytext=(140, edr_distance_lower+2),
                 arrowprops=dict(arrowstyle='<-'), fontsize=15)

# Amplitude value LOWER
plt.annotate(f'Amplitude = {edr_distance_lower:.0f} dB',
             xy=(80, edr_distance_lower),
             xytext=(150, edr_distance_lower-10),
                 arrowprops=dict(arrowstyle='<-'), fontsize=15)

## higher absorption rate, higher attenuation, quieter sooner
plt.plot(d, L_d_higher_absorption,color='dodgerblue',
          label='Higher Attenuation')

# abel='Higher Absorption ( = 0.25 dB/m)')
plt.scatter([80], [edr_distance_higher], color='dodgerblue', marker='*', s=400)
## EDr value HIGHER
plt.annotate(f'Minimum EDR = 80 m',
             xy=(80, edr_distance_higher),
             xytext=(80-10, edr_distance_higher-30),
                 arrowprops=dict(arrowstyle='<-'), fontsize=15)

# Amplitude value HIGHER
plt.annotate(f'Amplitude = {edr_distance_higher:.0f} dB',
             xy=(80, edr_distance_higher),
             xytext=(18, edr_distance_higher-15),
                 arrowprops=dict(arrowstyle='<-'), fontsize=15)
plt.axhline(0, linestyle='--')
plt.xlabel('Distance (m)', fontsize=20)
plt.ylabel('Sound Level (dB)', fontsize=20)
# changing the fontsize of yticks
plt.yticks(fontsize=15)
# changing the fontsize of yticks
plt.xticks(fontsize=15)
# plt.title('Sound Level over Distance with Different Absorption Coefficients')
plt.legend(fontsize=16)
plt.grid(False)

plt.savefig(os.path.join(figures_path,'SPL high res.png'), bbox_inches="tight", dpi=300)

plt.show()
```
